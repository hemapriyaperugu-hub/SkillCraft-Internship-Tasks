```python
msg=input("Enter the message: ")
shift=int(input("Enter the shift value: "))

encrypted_msg= ""

for letter in msg:
  if letter.isalpha():
    asc_value=ord(letter)
    new_value=asc_value+shift
    if letter.isupper():
      if new_value > ord('Z'):
        new_value = new_value-26
    if letter.islower():
      if new_value>ord('z'):
        new_value=new_value-26
    encrypted_msg=encrypted_msg+chr(new_value)
  else:
    encrypted_msg=encrypted_msg+letter


print("Encrypted Text:",encrypted_msg)

```
